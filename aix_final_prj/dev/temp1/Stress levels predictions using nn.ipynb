```python
import pandas as pd
full_data = pd.read_csv('synthetic_coffee_health_10000.csv')
full_data
```

```python
data = full_data.drop(['ID', 'Caffeine_mg', 'Sleep_Quality', 'Health_Issues', 'Heart_Rate'], axis=1)
data
```

```python
data = data[data['Gender'] != 'Other']
data = pd.get_dummies(data, columns=['Gender'], drop_first=True, dtype=int)
data
```

```python
from sklearn.preprocessing import LabelEncoder

encoders = {}
for col in ['Country', 'Occupation']: 
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    encoders[col] = le
data
```

```python
cols = list(data.columns)
i, j = cols.index("Stress_Level"), cols.index("Gender_Male")
cols[i], cols[j] = cols[j], cols[i]
data = data[cols]
data
```

```python
data["Stress_Level"] = pd.Categorical(
    data["Stress_Level"],
    categories=["Low", "Medium", "High"],
    ordered=True
).codes
data
```

```python
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split

class CoffeeDataset(Dataset):
    def __init__(self, data):
        self.X = data.drop('Stress_Level', axis=1).values
        self.y = data['Stress_Level'].values

    def __len__(self):
        return self.y.shape[0]

    def __getitem__(self, idx):
        return torch.tensor(self.X[idx], dtype=torch.float), torch.tensor(self.y[idx], dtype=torch.long)

train_data, test_val_data = train_test_split(data, test_size=0.4, stratify=data["Stress_Level"])
val_data, test_data = train_test_split(test_val_data, test_size=0.5, stratify=test_val_data["Stress_Level"])

train_dataset = CoffeeDataset(train_data)
val_dataset = CoffeeDataset(val_data)
test_dataset = CoffeeDataset(test_data)

batch_size = 512
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
import torch.nn as nn

class CoffeeModel(nn.Module):
    def __init__(self, n_countries, n_occupations):
        super().__init__()
        embed_dim = 10
        self.country_embed = nn.Embedding(n_countries, embed_dim)
        self.occupation_embed = nn.Embedding(n_occupations, embed_dim)

        self.fc = nn.Sequential(
            nn.Linear(embed_dim * 2 + 8, 20),
            nn.CELU(), # here is better than relu
            nn.Linear(20, 3)
        )

    def forward(self, x):
        country = self.country_embed(x[:, 1].long())
        occupation = self.occupation_embed(x[:, 8].long())

        x = torch.cat([country, occupation, x[:, 0].unsqueeze(1), x[:, 2:7], x[:, 8:]], dim=1)
        return self.fc(x)

class CoffeeModel(nn.Module):
    def __init__(self, n_countries, n_occupations):
        super().__init__()
        embed_dim = 10
        self.country_embed = nn.Embedding(n_countries, embed_dim)
        self.occupation_embed = nn.Embedding(n_occupations, embed_dim)

        self.fc = nn.Sequential(
            nn.BatchNorm1d(embed_dim * 2 + 8),
            nn.Linear(embed_dim * 2 + 8, 16),
            nn.BatchNorm1d(16),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(16, 8),
            nn.BatchNorm1d(8),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(8, 3)
        )

    def forward(self, x):
        country = self.country_embed(x[:, 1].long())
        occupation = self.occupation_embed(x[:, 7].long())

        x = torch.cat([country, occupation, x[:, 0].unsqueeze(1), x[:, 2:7], x[:, 8:]], dim=1)
        return self.fc(x)
```

```python
import torch
import torch.optim as optim
import copy
import numpy as np
from sklearn.metrics import f1_score

model = CoffeeModel(data['Country'].max() + 1, data['Occupation'].max() + 1)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.005)

min_loss = 100
patience = 5

for epoch in range(200):
    model.train()
    train_loss = 0
    for Xb, yb in train_loader:
        optimizer.zero_grad()
        preds = model(Xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        train_loss += loss.item() * Xb.size(0)

    train_loss /= len(train_loader.dataset)

    model.eval()
    val_loss = 0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for Xb, yb in val_loader:
            preds = model(Xb)
            loss = criterion(preds, yb)
            val_loss += loss.item() * Xb.size(0)
            
            all_preds.extend(preds.argmax(1).cpu().numpy())
            all_labels.extend(yb.cpu().numpy())

    
    val_loss /= len(val_loader.dataset)
    f1 = f1_score(all_labels, all_preds, average='macro')

    if val_loss <= min_loss:
        min_loss = val_loss
        patience = 5
        best_model_state = copy.deepcopy(model.state_dict())
    else:
        patience -= 1

    if patience == 0:
        break
        
    print(f"Epoch {epoch+1}: train_loss={train_loss:.4f}, val_loss={val_loss:.4f}, val_f1={f1:.4f}")

model.load_state_dict(best_model_state)
```

```python
import torch
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

model.eval()  # set model to evaluation mode

all_preds = []
all_labels = []

with torch.no_grad():
    for Xb, yb in test_loader:
        preds = model(Xb)

        predicted_classes = torch.argmax(preds, dim=1)
        all_preds.extend(predicted_classes.cpu().numpy())
        all_labels.extend(yb.cpu().numpy())

confusion_matrix(all_labels, all_preds)
```

```python
print(classification_report(all_labels, all_preds))
```
